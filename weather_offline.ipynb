```python
import os
import requests
from dotenv import load_dotenv
load_dotenv() #If this returns False, you can specify path to the .env inside the brackets
BASE_URL = "https://api.openweathermap.org/data/2.5/weather?"
API_KEY = os.getenv('WEATHER_API_KEY') #Best way. Can save many API keys in one file and reference them.
CITY = "Berlin"
url = f"{BASE_URL}appid={API_KEY}&q={CITY}"
response = requests.get(url)
values = response.json()
values
```
